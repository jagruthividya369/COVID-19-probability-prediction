```python
from statsmodels.tsa.vector_ar.var_model import VAR
diu_calc=[]
model = VAR(endog=pat_trans)
model_fit = model.fit()
yhat = model_fit.forecast(model_fit.y, steps=1)
diu_calc.append(yhat)
```

```python
pat_train=pd.read_excel("Train_dataset.xlsx",sheet_name='Train_dataset')
```

```python
pat_train_new=pat_train.drop(['Designation','Name','Insurance','salary'],axis='columns')
```

```python
ind=pat_train_new.select_dtypes(include=['object']).columns 
ind
```

```python
ind_num=pat_train_new.select_dtypes(include=['int64','float64']).columns
ind_num
```

```python
for i in ind:
    pat_train_new=pat_train_new[pat_train_new[i].notna()]
for i in ind_num:
    if i!='people_ID' and i!='Infect_Prob':
        pat_train_new[i].fillna(pat_train_new[i].mean(),inplace=True)
for i in ind:
    pat_train_new[i] = pat_train_new[i].astype('category').cat.codes
```

```python
pat_train_new.isna().sum()
```

```python
target=pat_train_new['Infect_Prob']
features=pat_train_new.drop(columns=['Infect_Prob'])
```

```python
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, random_state=42, test_size=.2)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
lr = linear_model.LinearRegression()
model = lr.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
train_pred=lr.predict(x_test)
print('Testing error_RootMeanSquareError', np.sqrt(mean_squared_error(y_test,train_pred)))
```
