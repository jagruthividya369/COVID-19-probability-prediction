```python
pat_data_new=pat_data.drop(['Designation','Name','Insurance','salary'],axis='columns')
```

```python
ind=pat_data_new.select_dtypes(include=['object']).columns 
ind
```

```python
ind_num=pat_data_new.select_dtypes(include=['int64','float64']).columns
ind_num
```

```python
pat_data_new.nunique()
```

```python
for i in ind:
    pat_data_new=pat_data_new[pat_data_new[i].notna()]
```

```python
pat_data_new[ind_num]
```

```python
from sklearn import preprocessing
for i in ind_num:
    if i!='people_ID' and i!='Infect_Prob':
        pat_data_new[i].fillna(pat_data_new[i].mean(),inplace=True)
```

```python
pat_data_new.isna().sum()
```

```python
pat_data_new.dtypes
```

```python
for i in ind:
    pat_data_new[i] = pat_data_new[i].astype('category').cat.codes
```

```python
target=pat_data_new['Infect_Prob']
features=pat_data_new.drop(columns=['Infect_Prob'])
```

```python
features.columns
```

```python
print(target.shape)
print(features.shape)
```

```python
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, random_state=42, test_size=.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
lr = linear_model.LinearRegression()
model = lr.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
x_test.shape
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
train_pred=lr.predict(x_train)
train_pred=lr.predict(x_test)
print('Testing error_RootMeanSquareError', np.sqrt(mean_squared_error(y_test,train_pred)))
```
